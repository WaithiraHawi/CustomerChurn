```python
import pandas as pd

# Load the uploaded dataset
file_path = '/content/Bank Customer Churn Prediction.csv'
df = pd.read_csv(file_path)

# Displaying basic info and preview
df_info = df.info()
df_head = df.head()

df_info, df_head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Basic descriptive statistics
desc_stats = df.describe()
```

```python
# Churn distribution
plt.figure(figsize=(5,4))
sns.countplot(x='churn', data=df, palette='coolwarm')
plt.title('Churn Distribution')
plt.xlabel('Churn (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

```python
# Churn rate by country
plt.figure(figsize=(6,4))
sns.barplot(x='country', y='churn', data=df, palette='viridis')
plt.title('Churn Rate by Country')
plt.ylabel('Average Churn Rate')
plt.show()
```

```python
# Age distribution by churn
plt.figure(figsize=(7,5))
sns.histplot(data=df, x='age', hue='churn', multiple='stack', palette='Set1', bins=20)
plt.title('Age Distribution by Churn Status')
plt.show()
```

```python
# Churn rate by active membership
plt.figure(figsize=(5,4))
sns.barplot(x='active_member', y='churn', data=df, palette='coolwarm')
plt.title('Churn Rate by Active Membership')
plt.show()

import pandas as pd
desc_stats
```
